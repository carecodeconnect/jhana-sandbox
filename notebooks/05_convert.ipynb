# Convert .m4a to .mp3

```python
import subprocess
import os
import ebooklib
from ebooklib import epub
from bs4 import BeautifulSoup
import fitz  # PyMuPDF
import os


```

```python

# Define the directory containing .m4a files
directory = "../data/input/audio/guided_meditations/audiodharma_dharmaseed/"

# Iterate over files in the specified directory
for filename in os.listdir(directory):
    if filename.endswith(".m4a"):
        # Construct full file path
        m4a_file = os.path.join(directory, filename)
        # Define the output .mp3 file path
        mp3_file = os.path.join(directory, filename[:-4] + '.mp3')
        
        # Check if the .mp3 file already exists
        if not os.path.exists(mp3_file):
            # Construct the ffmpeg command for conversion
            command = ['ffmpeg', '-i', m4a_file, '-acodec', 'libmp3lame', mp3_file]

            # Execute the conversion command
            subprocess.run(command)

            print(f"Converted {filename} to MP3 format.")
        else:
            print(f"Skipped {filename}, MP3 file already exists.")
            
print("Conversion complete.")

```

# Convert .epub to .txt

```python
def epub_to_txt(source_folder, target_folder):
    for file in os.listdir(source_folder):
        if file.endswith(".epub"):
            book_path = os.path.join(source_folder, file)
            book = epub.read_epub(book_path)
            content = []

            for item in book.get_items():
                if item.get_type() == ebooklib.ITEM_DOCUMENT:
                    soup = BeautifulSoup(item.get_content(), 'html.parser')
                    text = soup.get_text(separator=' ', strip=True)
                    content.append(text)

            txt_filename = os.path.splitext(file)[0] + '.txt'
            with open(os.path.join(target_folder, txt_filename), 'w', encoding='utf-8') as txt_file:
                txt_file.write('\n'.join(content))

source_folder = '../data/input/text/books'
target_folder = '../data/input/text/books_converted'
epub_to_txt(source_folder, target_folder)
```

# Convert .pdf to .txt

```python

def pdf_to_txt(source_folder, target_folder):
    for file in os.listdir(source_folder):
        if file.endswith(".pdf"):
            pdf_path = os.path.join(source_folder, file)
            doc = fitz.open(pdf_path)
            text = ''

            for page in doc:
                text += page.get_text()

            txt_filename = os.path.splitext(file)[0] + '.txt'
            with open(os.path.join(target_folder, txt_filename), 'w', encoding='utf-8') as txt_file:
                txt_file.write(text)

source_folder = '../data/input/text/books'
target_folder = '../data/input/text/books_converted'
pdf_to_txt(source_folder, target_folder)
```

# Convert .html to .txt

```python

# Define the source and target directories
source_dir = '../data/input/text/web_html'
target_dir = '../data/input/text/web_txt'

# Ensure the target directory exists
os.makedirs(target_dir, exist_ok=True)

# Iterate through all files in the source directory
for filename in os.listdir(source_dir):
    if filename.endswith('.html'):
        # Construct the full path to the source file
        source_file_path = os.path.join(source_dir, filename)

        # Read the HTML content
        with open(source_file_path, 'r', encoding='utf-8') as file:
            html_content = file.read()
        
        # Parse the HTML content and extract text
        soup = BeautifulSoup(html_content, 'html.parser')
        text_content = soup.get_text(separator='\n', strip=True)

        # Construct the target file path with .txt extension
        target_file_path = os.path.join(target_dir, os.path.splitext(filename)[0] + '.txt')

        # Write the text content to the target file
        with open(target_file_path, 'w', encoding='utf-8') as file:
            file.write(text_content)

# List the converted files to confirm
converted_files = os.listdir(target_dir)
converted_files
```
