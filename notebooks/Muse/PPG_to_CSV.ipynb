```python
"""
Record photoplethysmogram (PPG) signal from a Muse2 stream and save it in a CSV file

This script records data from an existing Muse LSL stream
(the stream should be started in a separate terminal or notebook)
"""

from muselsl import record

if __name__ == "__main__":

    seconds_to_record = 60
    # Note: an existing Muse LSL stream is required
    record(seconds_to_record, data_source="PPG")

    # Note: Recording is synchronous, so code here will not execute until the stream has been closed
    print('Recording has ended')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the DataFrame
df = pd.read_csv('PPG_recording_2024-03-06-10.35.39.csv')

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each channel separately
channels = ['PPG1', 'PPG2', 'PPG3']
for channel in channels:
    ax.plot(df['timestamps'], df[channel], label=channel, linewidth=1)

# Add labels and legend
ax.set_xlabel('Timestamps')
ax.set_ylabel('PPG Values')
ax.set_title('PPG Recording')
ax.legend()

# Show plot
plt.grid(True)
plt.show()
```
