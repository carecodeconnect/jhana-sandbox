```python
"""
Record EEG from a Stream and save it in a CSV file

This script records data from an existing Muse LSL stream
(the stream should be started in a separate terminal or notebook)
"""
from muselsl import record

if __name__ == "__main__":

    seconds_to_record = 60
    # Note: an existing Muse LSL stream is required
    record(seconds_to_record)

    # Note: Recording is synchronous, so code here will not execute until the stream has been closed
    print('Recording has ended')
```

```python
"""
Plot the recorded data as 4 separate time series
"""

import pandas as pd
import matplotlib.pyplot as plt

# Read the DataFrame
df = pd.read_csv('EEG_recording_2024-03-05-12.38.55.csv')

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each channel separately
channels = ['TP9', 'AF7', 'AF8', 'TP10']
for i, channel in enumerate(channels):
    ax.plot(df['timestamps'], df[channel] + i * 100, label=channel, linewidth=0.2)

# Set y-ticks and labels
ax.set_yticks(range(0, len(channels) * 100, 100))
ax.set_yticklabels(channels)

# Add labels and legend
ax.set_xlabel('Timestamps')
ax.set_ylabel('EEG Values')
ax.set_title('EEG Recording')
ax.legend()

# Show plot
plt.grid(True)
plt.show()
```
